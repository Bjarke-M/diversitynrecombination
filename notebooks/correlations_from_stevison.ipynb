```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the second tab of the Excel file into a DataFrame
df_tab2 = pd.read_excel('/home/bjarkemp/primatediversity/people/bjarkemp/diversitynrecombination/data/Supplementary_Tables_stevison_et_al_2015.xlsx', sheet_name=1, header=0)
print(df_tab2)
```

```python
# Rename columns for easier access
df_tab2.columns = ['Chromosome', 'hg18_Start', 'hg18_End', 'HapMap_Rate', 'HapMap_YRI_Rate', 'HapMap_CEU_Rate', 
                   '10YRI_Rate', '10CEU_Rate', 'Western_Chimp_Rate', 'Nigerian_Chimp_Rate', 'Nigerian_Orientation', 
                   'Bonobo_Rate', 'Bonobo_Orientation', 'Gorilla_Rate', 'Gorilla_Orientation']

# Filter out rows where Chromosome is NaN
df_tab2 = df_tab2.dropna(subset=['Chromosome'])
```

```python
df_tab2 = df_tab2.drop(index=0)
```

```python
df_tab2
```

```python

# Convert species columns to numeric, coercing errors to NaN
species_columns = ['HapMap_Rate', 'HapMap_YRI_Rate', 'HapMap_CEU_Rate', '10YRI_Rate', '10CEU_Rate', 
                   'Western_Chimp_Rate', 'Nigerian_Chimp_Rate', 'Bonobo_Rate', 'Gorilla_Rate']
df_tab2[species_columns] = df_tab2[species_columns].apply(pd.to_numeric, errors='coerce')

# Get unique chromosomes
chromosomes = df_tab2['Chromosome'].unique()
```

```python
df_tab2
```

```python

# Dictionary to store correlation results
correlation_results = {}

# Calculate Spearman's rank correlation for each chromosome
for chrom in chromosomes:
    chrom_data = df_tab2[df_tab2['Chromosome'] == chrom]
    corr_matrix = chrom_data[species_columns].corr(method='spearman')
    correlation_results[chrom] = corr_matrix

# Display correlation results for each chromosome
for chrom, corr_matrix in correlation_results.items():
    print(f"Correlation matrix for {chrom}:")
    print(corr_matrix)
    print("\n")
```

```python
# Calculate the average correlation for each species column
average_correlation = corr_matrix.mean()
print(average_correlation)
```

```python
# Filter the chrom_data DataFrame to include only the specified samples
selected_samples = ['HapMap_Rate', 'Western_Chimp_Rate', 'Nigerian_Chimp_Rate', 'Bonobo_Rate', 'Gorilla_Rate']
filtered_data = chrom_data[selected_samples]

# Calculate the correlation matrix
correlation_matrix = filtered_data.corr(method='spearman')

# Display the correlation matrix
print(correlation_matrix)
```

```python
correlation_matrix.mean()
```

```python
# Initialize dictionaries to store min and max correlations for each species
min_correlation = {}
max_correlation = {}

# Calculate min and max correlations for each species, excluding itself
for species in selected_samples:
    correlations = correlation_matrix[species].drop(species)
    min_correlation[species] = correlations.min()
    max_correlation[species] = correlations.max()

# Display the results
print("Minimum correlations for each species (excluding itself):")
print(min_correlation)
print("\nMaximum correlations for each species (excluding itself):")
print(max_correlation)
```
